**Data Processing for ETL:**

- Normalization
- Drop columns `Unnamed:0` and `Unnamed:0.1`
- Remove "fraud" string from `merchant`
- Drop duplicates
- Drop missing values
- Change datatype:
    - `trans_date_trans_time` -> timestamp
    - `amt` -> float
    - `dob` -> datetime (yyyy-mm-dd)
    - `unix_time` -> timestamp (unix format)

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True) # Drop the first column
df.drop_duplicates(inplace=True) # Drop duplicates 
df.dropna(inplace=True) # Drop duplicates 
```

```python
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
df["dob"] = pd.to_datetime(df["dob"])
df["unix_time"] = pd.to_datetime(df["unix_time"], origin='unix', unit='s')

# 2. Replace commas with dots and convert to numeric
comma_to_float_cols = ["amt", "lat", "long", "merch_lat", "merch_long"]
for col in comma_to_float_cols:
    df[col] = df[col].str.replace(",", ".").astype(float)
```

```python
df['merchant'] = df['merchant'].str.replace("fraud", "")
# 4. Strip spaces and lowercase categorical strings
cat_cols = ["merchant", "category", "first", "last", "gender", "street", "city", "state", "job"]
for col in cat_cols:
    df[col] = df[col].str.strip()
    df[col] = df[col].str.replace("_", " ")
```
