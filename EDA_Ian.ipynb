```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import scipy 
```

```python
# Load dataset
df = pd.read_csv('DatosPrueba_clean.csv')
df.head()
```

```python
df.info()
```

```python
# Visualisasi Fraud vs Non Fraud(Bar Chart)
plt.figure(figsize=(8, 6))
sns.countplot(x='is_fraud', data=df)
plt.title("Distribusi Fraud vs Non Fraud")
plt.xlabel("Distribusi")
plt.ylabel("Jumlah")
plt.xticks([0, 1], ['Non Fraud', 'Fraud'])
plt.show()
```
